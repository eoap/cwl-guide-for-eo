## Build the container supporting the How-to's

The `Dockerfile` adds `rasterio` to a GDAL base container image:

```bash
cat ../Dockerfile
```

Build the image with `podman` container engine:

```bash
podman build -t rio:latest ..
```

List the built image:

```bash
podman images
```

Test the container:

```bash
podman run --rm -it localhost/rio:latest rio --help
```
